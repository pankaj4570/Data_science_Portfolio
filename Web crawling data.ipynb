```python
nbcc_url= 'https://www.nbcnews.com/health/coronavirus'
the_hindu = 'https://www.thehindu.com/news/national/coronavirus-live-updates-march-23-2020/article31140072.ece'
```

```python
urls =[cnn_url, nbcc_url, the_hindu]
formats = ['html.parser', 'html.parser', 'html.parser']
tags = ['h2', 'h2', 'h2']
websites = ['cnn', 'nbcc', 'The_hindu']
# # response = requests.get(cnbc_url)
# # soup = BeautifulSoup(response.content,'xml')
# # print(soup.find_all('description').text)

# response = requests.get(the_hindu)
# soup = BeautifulSoup(response.content,'html.parser')
# for link in soup.find_all('h2'):
#     print (link.text)

# # urls =[cnn_url, cnbc_url]
# # formats = ['html.parser', 'xml']
# # tags = ['h2', 'description']
# # websites = ['cnn','cnbc']
```

```python
crawl_len = 0
for url in urls:
    print('crawling web page ........{}'.format(url))
    response = requests.get(url)
    soup = BeautifulSoup(response.content,formats[crawl_len])
    for link in soup.find_all(tags[crawl_len]):   
        if len(link.text.split(' ')) > 4:
            print('Headline is : {}'.format(link.text))
            Entities = []
            Entities=[(ent.text, ent.label_) for ent in nlp(link.text).ents]
            print('\tText :{}, Entity : {}'.format(ent.text, ent.label_))
    crawl_len+=1         
```

```python
crawl_len = 0
new_dict= []
for url in urls:
    response = requests.get(url)
    soup = BeautifulSoup(response.content,formats[crawl_len])
    
    for link in soup.find_all(tags[crawl_len]):
        if  len(link.text.split(' '))> 4 :
            Entities = []
            Entities=[(ent.text, ent.label_) for ent in nlp(link.text).ents]
            new_dict.append({'Website':websites[crawl_len], 'url':url, 'headline': link.text, 'entities': Entities })
    crawl_len = crawl_len+1         
```

```python
new_dict
```

```python
import pandas as pd
web_crawled__df = pd.DataFrame(new_dict)
```

```python
web_crawled__df['Website'].unique()
```

```python
web_crawled__df = pd.concat([web_crawled__df[['Website', 'url', 'headline']],web_crawled__df['entities'].apply(pd.Series)], axis =1)
```

```python
web_crawled__df.to_csv('web_crawled_data.csv', index=False, header= True,columns=['Website','url','headline'])
```
